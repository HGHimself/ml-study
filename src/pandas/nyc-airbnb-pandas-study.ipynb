```python
import pandas as pd
```

```python

airbb_data = pd.read_csv('../data/new-york-city-airbnb-open-data/AB_NYC_2019.csv')
```

```python
airbb_data.describe()
```

```python
airbb_data.price.describe()
```

```python
airbb_data_price_mean = airbb_data.price.mean()

airbb_data.price.map(lambda p: p - airbb_data_price_mean)
```

```python
(airbb_data.describe()).price[1]
```

```python
def remean_points(row):
    row.price = row.price - airbb_data_price_mean
    return row

airbb_data.apply(remean_points, axis='columns')

```
